```python
%matplotlib inline
import random
import matplotlib.pyplot as plt


# 
# figure
plt.figure(figsize = (20, 8), dpi=100)

# xy
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]

# 
x_ch = ["11{}".format(i) for i in x]
y_ticks = range(40)

# x,y
plt.xticks(x[::5], x_ch[::5])
plt.yticks(y_ticks[::5])

# 
plt.plot(x, y_shanghai, label='')
plt.show()
```
